```python

def prepare(mesh):
    mesh_g = tri.load_mesh(mesh)

   # Temporarily align the mesh with the global coordinate system
    A = mesh_g.principal_inertia_transform
    mesh_g.apply_transform(A)
    return (mesh_g) 
```

```python

#load the STL files
scapula_avg = 'PJ116_scapula_A_avg.stl' 
scapula_AMMR ='scapula_generique.stl'

mesh1= prepare(scapula_avg)
mesh2 =  prepare(scapula_AMMR)
```

```python
# Compute the maximum extents of each mesh
mesh1_max_extent = max(mesh1.bounds.max(axis=0) - mesh1.bounds.min(axis=0))
mesh2_max_extent = max(mesh2.bounds.max(axis=0) - mesh2.bounds.min(axis=0))

# Compute the scaling factor
scale_factor = mesh2_max_extent / mesh1_max_extent


# Scale the vertices of the larger mesh
mesh1.vertices *= scale_factor

# Save the rescaled mesh
# mesh2.export('path/to/rescaled_mesh2.stl')
```

```python
mesh1.visual.vertex_colors = [255, 0, 0, 255]
mesh2.visual.vertex_colors = [0, 0, 255, 255]
(mesh1 + mesh2).show()
```

```python
# Register mesh1 on mesh2
A = tri.registration.mesh_other(mesh1, mesh2, scale=True,trshold=10e-20, icp_first=200)
```

```python
# Apply the transformation to mesh1
mesh1.apply_transform(A[0])
```

```python
# Offset the transformed mesh1 to align with the original frame of mesh2
offset = mesh2.centroid - mesh1.centroid
mesh1.vertices += offset

(mesh1 + mesh2).show()
```

```python
# Save the transformed mesh1
mesh1.export('scapula_avg_reg_3.stl')
mesh2.export('scapula_AMMR_reg_3.stl')
```
